```python
list1 = []
for a1 in aa:
    event_name = a1['name']
    address = a1['venue']['address_1']
    city = a1['venue']['city']
    event_id = a1['id']

    b = client.GetGroupEventsAttendance(id=event_id, urlname='tordatascience')
    for x in b:
        attendees = x.member['name']
        file_save=[event_id,event_name, address,city,  attendees]
        list1.append(file_save)
        ad2 = pd.DataFrame(list1,columns=['event_id','event_name','event_location','city', 'attendees'])
```
